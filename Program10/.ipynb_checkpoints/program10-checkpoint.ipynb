```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mouse_viral_study.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
sns.scatterplot(x='Med_1_mL', y='Med_2_mL', hue='Virus Present', data=df, palette='seismic')
```

```python
sns.scatterplot(x='Med_1_mL', y='Med_2_mL', hue='Virus Present', data=df, palette='seismic')

x = np.linspace(0, 10, 100)
m = -1
b = 11
y = m*x + b
plt.plot(x, y, 'k')
```

```python
from sklearn.svm import SVC
```

```python
y = df['Virus Present']
X = df.drop('Virus Present', axis=1)
```

```python
model = SVC(kernel='linear', C=1000)
model.fit(X, y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='linear', C=0.005) # check as C moves closer to zero
model.fit(X, y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly', C=1,degree=2)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly', C=1,degree=2)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1,gamma='auto')
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1],'kernel':['linear','rbf']}
grid = GridSearchCV(svm,param_grid)
```

```python
# Note again we didn't split Train|Test
grid.fit(X,y)
```

```python
# 100% accuracy (as expected)
grid.best_score_
```
